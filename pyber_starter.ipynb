```python
%matplotlib notebook
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
df=pd.merge(city_data, ride_data, how="left", on=['city','city'])

# Display the data table for preview
df
```

```python
# Create datafrmes by type
dfUrban = df[df['type']=='Urban'] 
dfSuburban =df[df['type']=='Suburban']
dfRural =df[df['type']=='Rural']
```

## Bubble Plot of Ride Sharing Data

```python
UavgFare_pcity = dfUrban.groupby('city')['fare'].mean()
UtotRides_pcity = dfUrban.groupby('city')['ride_id'].count()
UtotDrivers_pcity = dfUrban.groupby('city')['driver_count'].mean()
SavgFare_pcity = dfSuburban.groupby('city')['fare'].mean()
StotRides_pcity = dfSuburban.groupby('city')['ride_id'].count()
StotDrivers_pcity = dfSuburban.groupby('city')['driver_count'].mean()
RavgFare_pcity = dfRural.groupby('city')['fare'].mean()
RtotRides_pcity = dfRural.groupby('city')['ride_id'].count()
RtotDrivers_pcity = dfRural.groupby('city')['driver_count'].mean()

#print(TotRides)
```

```python
# Build the scatter plots for each city types

plt.scatter(UtotRides_pcity, UavgFare_pcity, s=UtotDrivers_pcity*10, marker="o", c='Coral',alpha=0.6, edgecolors="black", label="Urban")
plt.scatter(StotRides_pcity, SavgFare_pcity, s=StotDrivers_pcity*10, marker="o", c='SkyBlue',alpha=0.6,edgecolors="black", label="Suburban")
plt.scatter(RtotRides_pcity, RavgFare_pcity, s=RtotDrivers_pcity*10, marker="o", c='Gold',alpha=0.6, edgecolors="black",label ="Rural")

# Incorporate a text label regarding circle size
plt.text(42,35,"Note:")
plt.text(42,33.5,"Circle size is relative to Drivers Per City")

# Create a legend
legend = plt.legend(loc="best", title="City Types", title_fontsize="large")
legend.legendHandles[0]._sizes=[30]
legend.legendHandles[1]._sizes=[30]
legend.legendHandles[2]._sizes=[30]

# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
#plt.title("$\mathbf{Pyber Ride Share Data Jan- Early May 2018}$", fontsize=14)  Bold title
plt.title("Pyber Ride Share Data Jan - Early May 2018", fontsize=15, fontweight=0, style='italic' )


# Save figure
# Show plot
plt.savefig("PyberBubble.png", bbox_inches='tight', pad_inches=0.5)
plt.show()

```

## Total Fares by City Type

```python
# Calculate Type, Fare, Rides and Driver Percents
TotDrivers= df.groupby('city')['driver_count'].mean().sum()
TotFares= df['fare'].sum()
TotRides=df['ride_id'].count()
TotFares_perCityType= df.groupby('type')['fare'].sum()

CityPercent_totFares=(TotFares_perCityType/TotFares)*100
CityPercent_totRides=(df.groupby('type')['ride_id'].count()/TotRides)*100
CityPercent_totDrivers=df.groupby(['city','type'])['driver_count'].mean().groupby('type').sum()

# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]
sizes = CityPercent_totFares
colors = ["Gold", "Skyblue", "Coral"]
explode = (0, 0, 0.1, )

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Fares By City Type")
plt.axis("equal")

# Save Figure
plt.savefig("PyberPieFares.png")
# Show Figure
plt.show()
```
